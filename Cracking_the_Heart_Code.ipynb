## Data and Methods

### Data Description
link: https://www.kaggle.com/datasets/alexteboul/heart-disease-health-indicators-dataset

The dataset for this project is derived from the **Behavioral Risk Factor Surveillance System (BRFSS)**, a publicly available health-related dataset collected annually by the Centers for Disease Control and Prevention (CDC). It includes responses from over 250,000 individuals on various health indicators, providing a rich source for predicting heart disease. Key characteristics of the dataset include:
- **Number of Records:** 253,680 (after cleaning)
- **Target Variable:** `HeartDiseaseorAttack` (binary classification: 1 = heart disease present, 0 = no heart disease)
- **Features:** 20+ variables covering demographics, lifestyle habits, and medical history, such as:
  - `BMI`: Body Mass Index
  - `Smoking`: Whether the respondent has smoked at least 100 cigarettes in their lifetime
  - `PhysicalActivity`: Engagement in physical activities in the last 30 days
  - `HighBP`: History of high blood pressure

The dataset presents a significant **class imbalance**, with only ~10% of respondents classified as having heart disease. Addressing this imbalance is a critical step in ensuring reliable model performance.

### Data Cleaning and Preprocessing

The raw dataset required extensive cleaning and preprocessing to ensure usability for machine learning:
1. **Handling Missing Values:** Removed rows with missing or invalid responses (e.g., "Don't know" or "Refused").
2. **Feature Engineering:**
   - Encoded categorical variables (e.g., Yes/No responses converted to 0/1).
   - Normalized continuous variables (e.g., BMI scaled appropriately).
3. **Addressing Class Imbalance:**
   - Applied **SMOTE (Synthetic Minority Oversampling Technique)** to oversample the minority class.
   - Experimented with **random undersampling** for balanced comparison.

### Machine Learning Methods

We focused on **supervised learning** for binary classification. The following machine learning models were implemented:
- **Baseline Models:**
  - Logistic Regression
  - K-Nearest Neighbors (KNN)
  - Decision Tree
- **Advanced Models:**
  - Random Forest
  - AdaBoost
  - XGBoost

### Experimental Setup

1. **Evaluation Metrics:**
   - Accuracy
   - Precision, Recall, and F1-Score
   - Area Under the Receiver Operating Characteristic Curve (AUC-ROC)
2. **Hyperparameter Optimization:**
   - Used **Grid Search** and **Randomized Search** with cross-validation for parameter tuning.
3. **Validation Strategy:**
   - 80/20 train-test split for initial testing.
   - 5-fold cross-validation to ensure robustness.

### Trade-offs
- **Decision Tree (DT)**:
  - Simple and interpretable; ideal for stakeholder communication.
  - Prone to overfitting; less robust than ensemble methods.

- **Random Forest (RF)**:
  - Most accurate model with excellent robustness due to ensemble learning.
  - Requires more computational resources; less interpretable compared to DT.

- **Why RF and DT?**
  - RF offers the best predictive accuracy and generalization capability.
  - DT provides intuitive insights into feature importance, facilitating actionable business decisions.

## Preventing Data Leakage

To ensure the validity and robustness of our model predictions, we took proactive measures to prevent data leakage during the analysis. Data leakage can occur when information from the test set influences the training process, leading to overly optimistic performance estimates. The following steps were implemented:

1. **Feature Engineering on Training Data Only**:
   - All transformations (e.g., scaling, encoding, and imputation) were applied only to the training data and subsequently applied to the test data to avoid information leakage.
   
2. **Data Splitting**:
   - The dataset was split into training and test sets before any feature engineering or oversampling techniques (e.g., SMOTE) were applied. This ensures the test data remains unseen during the training process.

3. **Cross-Validation**:
   - We used k-fold cross-validation on the training set to further mitigate the risk of overfitting and ensure that model hyperparameters were tuned without exposing the model to test data.

These steps ensure that the models evaluate performance on unseen data, reflecting their true predictive power in real-world applications.


```python
def evaluate_model(model, param_grid, X, y, model_name, n_iter=20, cv=5):
    """
    Evaluate a given model using RandomizedSearchCV for hyperparameter tuning,
    and return the evaluation results.
    """
    # Create a pipeline with a scaler and the classifier
    pipeline = Pipeline([
        ('scaler', StandardScaler()),  # StandardScaler for feature scaling
        ('classifier', model)
    ])

    # Hyperparameter tuning using RandomizedSearchCV
    random_search = RandomizedSearchCV(pipeline, param_grid, n_iter=n_iter, cv=cv, scoring='accuracy', random_state=42)
    random_search.fit(X_train, y_train)

    # Evaluate the best model on the test data
    best_model = random_search.best_estimator_
    y_pred_train = best_model.predict(X_train)
    y_pred_test = best_model.predict(X_test)

    # Store results
    results = {
        'model_name': model_name,
        'best_params': random_search.best_params_,
        'best_model': best_model,
        'train_classification_report': classification_report(y_train, y_pred_train, output_dict=True),
        'test_classification_report': classification_report(y_test, y_pred_test, output_dict=True),
        'train_confusion_matrix': confusion_matrix(y_train, y_pred_train),
        'test_confusion_matrix': confusion_matrix(y_test, y_pred_test),
        'train_roc_auc': roc_auc_score(y_train, best_model.predict_proba(X_train)[:, 1]),
        'test_roc_auc': roc_auc_score(y_test, best_model.predict_proba(X_test)[:, 1]),
    }

    # Print evaluation metrics
    print(f"{results['model_name']} Best Parameters: {results['best_params']}")
    print(f"{results['model_name']} Training Classification Report:\n", classification_report(y_train, y_pred_train))
    print(f"{results['model_name']} Test Classification Report:\n", classification_report(y_test, y_pred_test))

    # Plot confusion matrices
    plot_confusion_matrix(results['train_confusion_matrix'], f"{model_name} - Train")
    plot_confusion_matrix(results['test_confusion_matrix'], f"{model_name} - Test")

    return results
```

```python
# Evaluation
def extract_metrics(results):
    """
    Extract performance metrics from the results dictionary and create a summary DataFrame.

    Args:
        results (dict): Dictionary containing evaluation metrics for each model.

    Returns:
        pd.DataFrame: Consolidated metrics for all models.
    """
    metrics = []
    for model_name, result in results.items():
        train_report = result['train_classification_report']
        test_report = result['test_classification_report']

        metrics.append({
            'Model': model_name,
            'Train Accuracy': train_report['accuracy'],
            'Train Precision': train_report['weighted avg']['precision'],
            'Train Recall': train_report['weighted avg']['recall'],
            'Train F1-Score': train_report['weighted avg']['f1-score'],
            'Train ROC-AUC': result['train_roc_auc'],

            'Test Accuracy': test_report['accuracy'],
            'Test Precision': test_report['weighted avg']['precision'],
            'Test Recall': test_report['weighted avg']['recall'],
            'Test F1-Score': test_report['weighted avg']['f1-score'],
            'Test ROC-AUC': result['test_roc_auc']
        })
    return pd.DataFrame(metrics)
```

```python
def extract_decision_tree_splits(tree_model, feature_names):
    """
    Extract all split thresholds and weights from a trained Decision Tree model,
    skipping splits with zero or truncated weights.

    Args:
        tree_model (DecisionTreeClassifier): Trained Decision Tree model.
        feature_names (list): List of feature names corresponding to the dataset.

    Returns:
        pd.DataFrame: Table containing split variable, all thresholds, and weights.
    """
    tree = tree_model.tree_
    splits = {}

    for i in range(tree.node_count):
        # Check if it's a split node (not a leaf)
        if tree.children_left[i] != tree.children_right[i]:
            feature_idx = tree.feature[i]
            threshold = tree.threshold[i]
            samples_at_node = tree.weighted_n_node_samples[i]
            total_samples = tree.weighted_n_node_samples[0]

            # Compute the weight proportion, round it, and filter out zero values
            weight_proportion = round(samples_at_node / total_samples, 3)
            if weight_proportion == 0:
                continue

            feature_name = feature_names[feature_idx]
            if feature_name not in splits:
                splits[feature_name] = {
                    "Thresholds": [],
                    "Weights (Proportions)": []
                }

            # Append threshold and weight
            splits[feature_name]["Thresholds"].append(round(threshold, 3))
            splits[feature_name]["Weights (Proportions)"].append(weight_proportion)

    # Convert the results into a DataFrame
    split_data = []
    for feature, data in splits.items():
        if data["Thresholds"]:  # Ensure there are valid splits for the feature
            split_data.append({
                "Feature": feature,
                "Thresholds": data["Thresholds"],
                "Weights (Proportions)": data["Weights (Proportions)"]
            })

    return pd.DataFrame(split_data)
```
